```python
import PIL.Image
import pilgram
from matplotlib import pyplot as plt
```

```python
img = PIL.Image.open('girl.jpg')
```

```python
plt.imshow(img)
plt.show()
```

```python
img_lofi = pilgram.lofi(img)
plt.imshow(img_lofi)
plt.show()
```

```python
img_inkwell = pilgram.inkwell(img)
plt.imshow(img_inkwell)
plt.show()
```

```python
img_kelvin = pilgram.kelvin(img)
plt.imshow(img_kelvin)
plt.show()
```

```python
img1 = pilgram.css.saturate(img, 2)
plt.imshow(img1)
plt.show()
```

```python
img2 = pilgram.css.saturate(img, 0.5)
plt.imshow(img2)
plt.show()
```

```python
img3 = pilgram.css.hue_rotate(img, 20)
plt.imshow(img3)
plt.show()
```

```python
img4 = pilgram.css.hue_rotate(img, -200)
plt.imshow(img4)
plt.show()
```

```python
img1 = PIL.Image.open('girl.jpg')
img2 = PIL.Image.open('cc.png')
img_blended = pilgram.css.blending.color(img1, img2)
plt.imshow(img_blended)
plt.show()
```

```python
img1 = PIL.Image.open('girl.jpg')
img2 = PIL.Image.open('cc.png')
img_darken = pilgram.css.blending.darken(img1, img2)
plt.imshow(img_darken)
plt.show()
```

```python
img1 = PIL.Image.open('girl.jpg')
img2 = PIL.Image.open('cc.png')
img_lighten = pilgram.css.blending.lighten(img1, img2)
plt.imshow(img_lighten)
plt.show()
```

```python
img1 = PIL.Image.open('girl.jpg')
img2 = PIL.Image.open('cc.png')
img_difference = pilgram.css.blending.difference(img1, img2)
plt.imshow(img_difference)
plt.show()
```

```python
img1 = PIL.Image.open('girl.jpg')
img2 = PIL.Image.open('cc.png')
img_multiply = pilgram.css.blending.multiply(img1, img2)
plt.imshow(img_multiply)
plt.show()
```
